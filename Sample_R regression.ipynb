### Univariate multinominal logistic regression

```r
var.to.use=c(4, 6, 8, 10, 12, 18, 16, #7
              43, 20, 22, 44, #4
             27:30, 34, 32, 36) #7
exp.coef=list()
exp.ci=list()
p=list()

i=1
i=18
#data[,var.to.use[i]]

data$adh_group <- relevel(factor(data$adh_group), ref = "adhere") 
```

```r
for(i in 1:18){
  
  if(i %in% c(1, 7,8))
  {res=multinom(data=data,adh_group~data[,var.to.use[i]])
  exp.coef[[i]]=exp(coef(res))
  exp.ci[[i]]= exp(confint(res, level=0.95))
  z = summary(res)$coefficients/summary(res)$standard.errors
  p[[i]] = (1-pnorm(abs(z), 0, 1))*2}
  
  else{
    res=multinom(data=data,adh_group~factor(data[,var.to.use[i]]))
    exp.coef[[i]]=exp(coef(res))
    exp.ci[[i]]= exp(confint(res, level=0.95))
    z = summary(res)$coefficients/summary(res)$standard.errors
    p[[i]] = (1-pnorm(abs(z), 0, 1))*2   
  } 
}


names(exp.coef)=colnames(data)[c(4, 6, 8, 10, 12, 18, 16, #7
                                 43, 20, 22, 44, #4
                                 27:30, 34, 32, 36)]
names(exp.ci)=colnames(data)[c(4, 6, 8, 10, 12, 18, 16, #7
                               43, 20, 22, 44, #4
                               27:30, 34, 32, 36)]
names(p)=colnames(data)[c(4, 6, 8, 10, 12, 18, 16, #7
                          43, 20, 22, 44, #4
                          27:30, 34, 32, 36)]
```

### Multivariate Multinominal 

```r
data$gender = factor(data$gender)
data$marital_status = factor(data$marital_status)
data$residence = factor(data$residence)
data$income = factor(data$income)
data$education_1 =factor(data$education_1)
data$econ = factor(data$econ)
data$disorder = factor(data$disorder)
data$activity_limit =factor(data$activity_limit)
data$depression = factor(data$depression)
data$conversation =factor(data$conversation)
data$understanding =factor(data$understanding)
data$curiosity = factor(data$curiosity)
data$decision = factor(data$decision)
data$unmet =factor(data$unmet)
data$usual_care = factor(data$usual_care)
data$health_care=factor(data$health_care)
```

```r
ulti_all<- multinom(adh_group ~ 
                       age + gender + marital_status + residence
                     +data$income + data$education_1 + data$econ + health_verse 
                     #patient-related
                     
                     + comorbid + data$disorder
                     + data$activity_limit + data$depression # condition-related
                     
                     +data$conversation + data$understanding + data$curiosity + data$decision
                     + data$unmet + data$usual_care + data$health_care, data=data) # healthcare team system 
```
