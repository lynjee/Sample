```python
# DM 진단여부 
DM = []
for i in range(0,85040):
    if ( cd['CDNUM'][i] == 2 ) & ( cd['CD3_2'][i] == 1):
        DM.append(1)
    else:
        DM.append(0)

cd['DM'] = DM

DM = cd.groupby('HPID')['DM'].sum()
DM = pd.DataFrame(DM)
DM.value_counts()
```

```python
# HL 진단여부 
HL = []
for i in range(0,85040):
    if ( cd['CDNUM'][i] == 3 ) & ( cd['CD3_2'][i] == 1):
        HL.append(1)
    else:
        HL.append(0)
        
cd['HL'] = HL

HL = cd.groupby('HPID')['HL'].sum()
HL = pd.DataFrame(HL)
HL.value_counts()
```

```python
# IHD 진단여부 
IHD = []
for i in range(0,85040):
    if ( cd['CDNUM'][i] == 6 ) & ( cd['CD3_2'][i] == 1):
        IHD.append(1)
    else:
        IHD.append(0)
        
cd['IHD'] = IHD

IHD = cd.groupby('HPID')['IHD'].sum()
IHD = pd.DataFrame(IHD)
IHD.value_counts()
```

```python
# CVD 진단여부 
CVD = []
for i in range(0,85040):
    if ( cd['CDNUM'][i] == 7 ) & ( cd['CD3_2'][i] == 1):
        CVD.append(1)
    else:
        CVD.append(0)
        
cd['CVD'] = CVD

CVD = cd.groupby('HPID')['CVD'].sum()
CVD = pd.DataFrame(CVD)
CVD.value_counts()
```

```python
md = []

for i in range(0,85040):
    if ( cd['comorbid '][i] ==1 ) & ( cd['CD3_2'][i]==1 ):
        md.append(1)
    else:
        md.append(0)
        
cd['comorbid'] = md
comorbidity = cd.groupby('HPID')['comorbid'].sum()
comorbidity = pd.DataFrame(comorbidity)
comorbidity.value_counts()
```

```python
# 절대 헷갈리지 말것: 이 부분에서 의사 처방 없으면 -1로 t18cd의 행을 날아가는 건 맞음
# 복약불순응 원인분석해보면 reason = df & cd file을 df HPID 기준으로 부분집합 추출한 것이기 때문에 cd.loc후 유효한 data가 1개라도 있던 HPID 기준으로 t18cd 살펴보면 
# 의사 약 처방 없는 경우가 부분적으로 섞여있더라도 살아나고
# 모든 질환에 대해 무응답 or 의사 약 처방 없거나 등 -1로 설정해서 loc.shape해서 한 사람에 대한 데이터가 다 사라진 경우는 원인 분석 때 안 나타남

adhere = [] 

list_non = [1,2,3,4,5,6]


for i in range(0,85040):
    if (cd['CD6'][i] in list_non ): # 주관적 이유로 약물 복용 안 하는 사람들 -> 0
        adhere.append(0)
    elif ( (cd['CD6'][i] == -1)  & (cd['CD9'][i] == 1) ): #5 -> (1) : 약물 복용방법대로 복용 ->1
        adhere.append(1)
    elif ( (cd['CD6'][i] == -1) & (cd['CD9'][i] == 2) ): #5 -> (2) : 약물 복용방법대로 복용하지 않음 ->0
        adhere.append(0)    
    elif ( (cd['CD6'][i] == -1) & (cd['CD9'][i] == -1) ): # 없음 
        adhere.append(-1) 
    elif ( (cd['CD6'][i] == 7) ): # 의사 처방 없어서 약 복용 없는 경우 -> 결측
        adhere.append(-1) 
    elif ( (cd['CD6'][i] == -6) ): # 만성질환 없는 경우 -> 결측 
        adhere.append(-1) 
    elif ( (cd['CD6'][i] == -9) ): # 무응답 -> 결측
        adhere.append(-1) 
    else:
        adhere.append(-1) # 기타 다른 값 -> 결측
    
adhere = pd.DataFrame(adhere)
#adherence
```
